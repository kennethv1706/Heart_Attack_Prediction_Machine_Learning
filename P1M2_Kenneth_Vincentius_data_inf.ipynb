```python
# import libraries
import pandas as pd
import numpy as np
import pickle
import random
```

```python
with open('full_process.pkl', 'rb') as file:
  full_process = pickle.load(file)
```

```python
#membuat data dummy untuk data inferences
np.random.seed(3)
n = 100
oldpeak = np.random.uniform(0,6.2,n)
chol = np.random.uniform(126,564,n)
thall =[0, 1, 2, 3]
thall = random.choices(thall, k=n)
cp = [0, 1, 2, 3]
cp = random.choices(cp, k=n)
exng = [0, 1]
exng = random.choices(exng, k=n)
sex= [0, 1]
sex = random.choices(sex, k=n)
caa= [0, 1, 2, 3,4]
caa = random.choices(caa, k=n)
slp= [0, 1, 2]
slp = random.choices(slp, k=n)
```

```python
data_inf = pd.DataFrame({
   'oldpeak' : oldpeak,
   'chol' : chol,
   'thall' : thall ,
   'cp' : cp,
   'exng' : exng ,
   'sex' : sex,
   'caa' : caa,
   'slp' : slp
})

data_inf
```

```python
y_pred_inf = full_process.predict(data_inf)
y_pred_inf
```

```python
# Mengkonversi ke dataframe

y_pred_inf_df = pd.DataFrame(y_pred_inf, columns=['Heart Attack - Prediction'])
y_pred_inf_df
```

```python
#mengconcat data inference dan hasil predict
tabel_inference = pd.concat([data_inf, y_pred_inf_df], axis=1)
tabel_inference
```

## Conceptual Problems

1. Jelaskan latar belakang adanya bagging dan cara kerja bagging !

Bagging dibuat untuk mengatasi masalah overfitting dan meningkatkan kinerja model. Baging muncul karena saat kita mengerjakan machine learning kita memiliki dataset yang kurang bagus sehingga menghasilkan model yang kurang baik sehingga bagging dibuat untuk mengatasi masalah tersebut.

Cara kerjanya : Bagging pertama-tama membuat beberapa set data latihan baru (bootstrap samples) dengan mengambil sampel acak dengan penggantian dari dataset asli . Lalu model difit ke booststrap sample , lalu hasil prediksi dari semua model dihasilkan dan kemudian diagregasi. Hasil prediksi digabungkan dengan cara mengambil rata-rata(untuk regresi) dan voting (untuk klasifikasi)

2. Jelaskan perbedaan cara kerja algoritma Random Forest dengan algoritma boosting yang Anda pilih !

Random Forest : menggunakan bagging untuk membangun beberapa pohon keputusan independen dengan pemilihan fitur acak, dan hasil prediksinya diambil berdasarkan rata-rata (regresi) atau mode (klasifikasi) dari pohon-pohon tersebut untuk mengurangi overfitting.

Boosting : teknik ensemble di mana setiap model berikutnya berusaha untuk mengoreksi kesalahan model sebelumnya dengan memberikan bobot lebih kepada data yang salah diprediksi, sehingga secara bertahap meningkatkan performa keseluruhan model.

3. Jelaskan apa yang dimaksud dengan Cross Validation !

Cross-validation adalah sebuah teknik yang digunakan dalam pembangunan model machine learning untuk mengevaluasi kinerja model dengan cara membagi dataset menjadi beberapa subset atau fold. Tujuannya adalah untuk menghindari overfitting dan memberikan tahu model mana yang akan memberikan hasil prediksi yang lebih baik
